```python
import os
import shutil

# Path to the folder containing the AVI files
source_folder = r"C:\Users\Comsats\Desktop\share\NTURGB120\nturgb+d_ir_003"

# Path to the folder where you want to save the extracted files
destination_folder = r"D:\NTU RGB Dataset\Extracted classes S003"

# Ensure the destination folder exists, create it if not
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Iterate through all files in the source folder
for i in range(1,61):
    for filename in os.listdir(source_folder):
    # Check if the file is an AVI file and contains "A001" in its title
        if filename.endswith(".avi") and f"A{str(i).zfill(3)}" in filename:
        # Construct the full path of the source file
            source_file = os.path.join(source_folder, filename)


         # Create folders for each class from A001 to A060 within the destination folder
            class_folder_name = f"class A{str(i).zfill(3)}"
            class_folder_path = os.path.join(destination_folder, class_folder_name)
            os.makedirs(class_folder_path, exist_ok=True)


        # Construct the full path of the destination file
            destination_file = os.path.join(class_folder_path, filename)


        # Copy the file from the source folder to the destination folder
            shutil.copyfile(source_file, destination_file)

        # Move the file to the corresponding class folder within the destination folder
#         shutil.move(source_file, destination_file)

        # Print a message indicating the file has been copied
            print(f"File '{filename}' copied to '{class_folder_path}'")

print("Extraction complete!")
```
